# ディープラーニングとは
機械学習モデルは入力データを意味ある出力に変換する。機械学習モデルの本質は入力データから有益な表現への変換を見つけ出すことにある。探す変換(計算)の集合を **仮設空間** と呼ばれ、機械学習は仮設空間から最適な変換を探す。

ディープラーニングは変換を層で表現し何層にも重ねる事を指す。他のアプローチでは1～2層に絞る傾向があり (**表層学習** または **シャローラーニング**)、ディープラーニングは変換を何層にも重ねる事で最終的な結果に関する情報を少しずつ増やす動きをする。モデルには以下要素があり、訓練ループ (training loop) を通して重みが学習される。

* 重み・パラメータ (weight / parameter)
* 損失関数・目的関数 (loss function / objective function)
* オプティマイザ (optimizer)

## 従来の機械学習
従来は表層学習とされる手法として以下が使われていた。

* 確率モデリング: ナイーブベイズアルゴリズム
* 初期のニューラルネットワーク
* カーネル法:  
  SVM。単純な分類問題では性能が高いが、大規模データセットへのスケーリングが難しかった。  
  表層学習の一種のため、 **特徴エンジニアリング** が必要
* 決定木:  
  決定木本体の他、ランダムフォレスト(RF)や勾配ブースティング木(GBM)がある。  
  RFやGBMノイズに頑健であるため、使いやすい学習器としてよく使われる

## 表層学習との比較
深層学習の特徴は **特徴エンジニアリングが自動化** される点にある。複数層からなる変換を通して複雑な入力を有益な表現に変換する。深層学習では複数層からなるニューラルネットワークが使われる。これはニューラルネットワークがモデルの多層に渡る変換部分を効率的に同時学習できる事による。長らくニューラルネットワークは多層モデルを効率的に訓練することができず、SVMなどの表層学習に対抗できなかった。近年の進歩によって現在は数千層の積み重ねの訓練も可能になっている。

* 活性化関数の改善
* 重みを初期化する方法の改善
* RMSProp/Adm といった最適化手法の改善
* バッチ正規化、残差接続、dw畳み込み...

現在の情勢では、表層学習問題へは勾配ブースティングマシン、知覚問題では深層学習がよく使われる。Kaggleの機械学習コンペではツールとしてはこれらの実装にXGBoost, Kerasがよく使われる。

## 深層学習の特徴
深層学習にはAI革命と呼ぶに相応しい幾つかの特徴がある。

* 単純さ  
  特徴エンジニアリングの必要性を無くし、複雑で脆く、技術力に大きく依存するパイプラインを単純なE2E訓練が可能なモデルに置き換えられる
* スケーラビリティ  
  GPU/TPUでの並列処理に非常に適しており、また訓練は任意のサイズのデータセットで行うことが出来る
* 多様性と再利用可能性  
  多くの学習器は毎回学習をやり直すが、深層学習では追加データで継続的に訓練が可能 (オンライン学習)。また、学習済みモデルを他の目的へも利用できる (転移学習)。
